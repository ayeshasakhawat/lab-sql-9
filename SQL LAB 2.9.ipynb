```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
query = '''SELECT COUNT(DISTINCT(last_name))
FROM sakila.actor;'''
data = pd.read_sql_query(query, engine)
data.head()
```

```python

query1 = '''SELECT *,
CASE 
WHEN DATE_FORMAT(rental_date,'%%W') = 'Saturday' then 'Weekend'
WHEN DATE_FORMAT(rental_date,'%%W') = 'Sunday' then 'Weekend'
ELSE 'Weekday'
END AS 'column_day_type'
FROM rental;'''
data1 = pd.read_sql_query(query1, engine)
data1.head()
```

```python
# connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
# engine = create_engine(connection_string)
query2 = '''SELECT *
FROM sakila.film
WHERE title LIKE '%%ARMAGEDDON%%';'''
data2 = pd.read_sql_query(query2, engine)
data2.head()
```

```python

query3 = '''SELECT title, length
FROM sakila.film
GROUP BY title, length
ORDER BY length DESC;'''
data3 = pd.read_sql_query(query3, engine)
data3.head()
```

```python

query4 = '''SELECT COUNT(title)
FROM sakila.film
WHERE special_features LIKE '%%Behind the Scenes%%';'''
data4 = pd.read_sql_query(query4, engine)
data4.head()
```

```python

query5 = '''SELECT rating, AVG(length)
FROM sakila.film
GROUP BY rating
HAVING AVG(length) > 120;'''
data5 = pd.read_sql_query(query5, engine)
data5.head()
```

```python

query6 = '''SELECT DENSE_RANK() OVER(ORDER BY length DESC) as 'Rank', title, length
FROM film
WHERE length != 0 and length IS NOT NULL;'''
data6 = pd.read_sql_query(query6, engine)
data6.head()
```
